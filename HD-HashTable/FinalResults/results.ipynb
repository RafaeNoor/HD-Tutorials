```python
import matplotlib.pyplot as plt
```

```python
num_reads = [30, 40, 50, 100, 150, 200]
total_accuracy = [88.16, 85.75, 85.75, 79.50, 74.57, 72.50]

plt.scatter(num_reads, total_accuracy, color='blue', marker='o', label='Overall Accuracy')
plt.title('Number of Reads in Dataset vs. Overall Accuracy')
plt.xlabel('Number of Reads')
plt.ylabel('Overall Accuracy')
plt.grid(True)
plt.legend()

plt.show()
```

```python
num_reads = [30, 40, 50, 100, 150, 200]
sensitivity = [87.78, 84.00, 85.50, 87.00, 91.33, 88.50]

# Sensitivity is also known as recall or true positive rate
plt.scatter(num_reads, sensitivity, color='blue', marker='o', label='Sensitivity')
plt.title('Number of Reads in Dataset vs. Sensitivity')
plt.xlabel('Number of Reads')
plt.ylabel('Sensitivity')
plt.grid(True)
plt.legend()

plt.show()
```

```python
# Data
num_reads = [30, 40, 50, 100, 150, 200]
specificity = [88.50, 87.50, 86.00, 72.00, 62.00, 56.50]

# Specificity is also known as true negative rate
plt.scatter(num_reads, specificity, color='blue', marker='o', label='Specificity')
plt.title('Number of Reads in Dataset vs. Specificity')
plt.xlabel('Number of Reads')
plt.ylabel('Specificity')
plt.grid(True)
plt.legend()

plt.show()
```
